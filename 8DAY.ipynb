```python
import numpy as np
from statistics import mode

data = [10,20,20,30,40]

print("mean : ",np.mean(data))
print("median : ",np.median(data))
print("mode : ",mode(data))
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name ": ["Amit","Ravi","Neha"," Raman","Rajat"],
    "Marks": [80,70,90,85,70]
})

print(df)
print(df["Marks"].var())
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
print(df.Marks.isnull())
print(df.isnull())
print(df.isnull().sum())
```

```python
print(df.sum())
print(df["Marks"].sum())
```

```python
print(df["Marks"].mean())
print(df["Marks"].median())
print(df["Marks"].mode())
print(df["Marks"].std())
print(df["Marks"].var())
print(df["Marks"].sum())
print(df["Marks"].min())
print(df["Marks"].max())
```

```python
data = [10,20,30,40,50]
print(np.percentile(data,25))
print(np.percentile(data,50))
print(np.percentile(data,75))
print(np.quantile(data,0.25))
print(np.quantile(data,0.50))
print(np.quantile(data,0.75))
```

```python
data = [10,20,6,40,50]
print(np.min(data))
print(np.argmin(data))
print(np.max(data))
print(np.argmax(data))
```

```python
data = [1,2,3]
print(np.cumsum(data)) # ye sum karta hai data ho one by one
print(np.cumprod(data)) # ye multiply karta hai data ho one by one
```
